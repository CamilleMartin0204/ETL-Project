```python
import pymongo
from splinter import Browser
from bs4 import BeautifulSoup
import os
from datetime import datetime
import pandas as pd
import platform
import pdb
import time
```

```python
def access_db(db_name, collection_name):
    conn = 'mongodb://localhost:27017'
    client = pymongo.MongoClient(conn)
    db = client[db_name]
    collection = db[collection_name]
    return db, collection
#query db, return db connection and reference to collection
def sort_data(collection): 
#     db, collection = access_db('ytube', 'trending_songs')
#     collection.find(), returns cursor 
    songs = collection.find().sort('views', pymongo.DESCENDING)
    return songs
def data_to_csv(rows):
    df = pd.DataFrame(list(rows))
    #pdb.set_trace()
#     df['views'] = pd.to_numeric(df['views'].str[:-6]) #transform(lambda x: int(x.strip(' views')))    
    df.to_csv('ETL_Project_output.csv', index = False)
    df.head()
def new_main():
    import requests
    db, collection = access_db('ytube', 'trending_songs')
    r = requests.get('https://www.youtube.com/feed/trending?bp=4gIuCggvbS8wNHJsZhIiUExGZ3F1TG5MNTlhbVBud2pLbmNhZUp3MDYzZlU1M3Q0cA%3D%3D')
    # views
    soup = BeautifulSoup(r.text, "html.parser")
    listings = soup.find_all('a', class_='yt-simple-endpoint style-scope ytd-video-renderer')
    views = soup.find_all('div', {'class': 'yt-lockup-meta'})#.find_all('li')[-1].text
    all_views = [v.find_all('li')[-1].text.strip(' views') for v in views]
    
    # titles
    titles = soup.find_all('div', {'class': 'yt-lockup-byline'})
    all_titles = [t.text for t in titles]
    dicts = []
    for order, (title, view) in enumerate(zip(all_titles, all_views)):
        scrape_dict = {'title':title, 'views': view,
         'datetime':datetime.now(), 'order':order}
        dicts.append(scrape_dict)
    collection.insert_many(dicts)
    sorted_list = sort_data(collection)
    data_to_csv(sorted_list)
    # Evan added this so we can use the dataframe instead of the csv to check results
    df = pd.DataFrame(dicts)
    return df
```

```python
df = new_main()
```

```python
df['views']=df.views.str.replace(',','')
df['views']=pd.to_numeric(df['views'])
df=df.sort_values(by='views', ascending=False)
def data_to_csv(dataframe):
    df.to_csv('ETL_Project_output.csv', index = False)
    df.head() 
data_to_csv(df)
```
